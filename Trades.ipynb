```python
import pandas as pd
pd.set_option("display.max_colwidth", 10000)
pd.options.mode.chained_assignment = None  # to not make too many copies

import openpyxl
```

```python
folder = r'C:\Users\tetyana.loskutova\Downloads\PC' + "\\"
```

```python
file_name = r"trade_2018-05-19 14_01_30.213647.xlsx"
```

```python
folder+file_name
```

```python
df_tr1 = pd.read_excel(folder + file_name)
```

```python
df_tr1
```

```python
df_luno_btc = df_tr1.loc[(df_tr1["exchange"] == 'LUNO') & (df_tr1["currency"] == 'BTC'), ("datetime","exchange","currency", 'base_currency', 'ask','volume', 'last_trade')]
```

```python
df_luno_btc.rename({'ask': 'ask_luno_btc', 'volume':'volume_luno_btc', 'last_trade':'last_trade_luno_btc'}, axis='columns', inplace=True)
```

```python
df_luno_btc
```

```python
df_luno_eth = df_tr1.loc[(df_tr1["exchange"] == 'LUNO') & (df_tr1["currency"] == 'ETH'), ("datetime","exchange","currency", 'base_currency', 'ask','volume', 'last_trade')]
```

```python
df_luno_eth.rename({'ask': 'ask_luno_eth', 'volume':'volume_luno_eth', 'last_trade':'last_trade_luno_eth'}, axis='columns', inplace=True)
```

```python
df_luno_eth
```

## Fixer

```python
df_fixer_eur = df_tr1.loc[(df_tr1["exchange"] == 'FIXER') & (df_tr1["currency"] == 'EUR'), ("datetime","exchange","currency", 'base_currency', 'ask','volume', 'last_trade')]
```

```python
df_fixer_eur.rename({'ask': 'ask_fixer_eur', 'volume':'volume_fixer_eur', 'last_trade':'last_trade_fixer_eur'}, axis='columns', inplace=True)
```

```python
df_fixer_eur
```

```python
df_fixer_zar = df_tr1.loc[(df_tr1["exchange"] == 'FIXER') & (df_tr1["currency"] == 'ZAR'), ("datetime","exchange","currency", 'base_currency', 'ask','volume', 'last_trade')]
```

```python
df_fixer_zar.rename({'ask': 'ask_fixer_zar', 'volume':'volume_fixer_zar', 'last_trade':'last_trade_fixer_zar'}, axis='columns', inplace=True)
```

```python
df_fixer_zar
```

# Merging

```python
df_tr2 = pd.merge(left=df_luno_btc, right=df_luno_eth, on='datetime', how='outer', sort=True)
```

```python
df_tr2
```

```python
df_tr3 = pd.merge(left=df_fixer_zar, right=df_fixer_eur, on='datetime', how='outer', sort=True)
```

```python
df_tr3
```

```python
df_tr4 = pd.merge(left=df_tr2, right=df_tr3, on='datetime', how='outer', sort=True)
```

```python
df_tr4
```

# Export results

```python
excel_writer = pd.ExcelWriter(folder + 'Export_trades.xlsx')
```

```python
df_tr4.to_excel(excel_writer)
```

```python
excel_writer.save()
```
