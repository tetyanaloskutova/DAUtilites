```python
from scipy.stats import pearsonr
import pandas as pd

def calculate_pvalues(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = round(pearsonr(df[r], df[c])[1], 4)
    return pvalues
```

## Improving the model

Because GDP is a difficult to change factor, let's build a model while normalising for GDP. The good thing is that in this case, small but significnt  relationship appears between the count of cases and HDI,

```python
df_hdi_with_all['Calc Count'] = lm_count.predict(df_hdi_with_all[['gdp','pm']])
```

```python
df_hdi_with_all['Calc Score']  = lm_score.predict(df_hdi_with_all[['rt','cti']])
```

```python
#Normalize actual GDP: count, ra, scope, cti, rt were found related and therefor excluded from the model
df_hdi_with_all['NormCount'] = df_hdi_with_all['Count'].astype(float)/df_hdi_with_all['gdp']*df_hdi_with_all['gdp'].max()
df_hdi_with_all['NormCTI'] = df_hdi_with_all['cti']/df_hdi_with_all['gdp']*df_hdi_with_all['gdp'].max()
df_hdi_with_all['NormRT'] = df_hdi_with_all['rt']/df_hdi_with_all['gdp']*df_hdi_with_all['gdp'].max()
```

```python
# Score works but we cannot use it because it is dependent on GDP
#lm_norm_count = smf.ols(formula='NormCount ~ score-1', data=df_hdi_with_all).fit()
lm_norm_count = smf.ols(formula='NormCount ~   total_inf   -1', data=df_hdi_with_all).fit()
#pm + score
lm_norm_count.summary()
```

However, the coefficients are so small and R-squared indicated that only about a third of the data set could be explained. Let's plot the relationship betweeb score and norm count to see what could be the proble,

```python
# Create legend for the initial IV for count and score
df_hdi_with_all['Legend'] = df_hdi_with_all.apply(lambda row: 'pm %.1f<br>cti %.1f<br>rt %.1f'%(row['pm'], row['cti'], row['rt']), axis=1)
```
